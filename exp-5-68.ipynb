### Expt.-5: Disease Prediction using Naive Bayes and Neural Network with Comparison of Classifiers 

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score , precision_score,recall_score ,f1_score,roc_auc_score,confusion_matrix , classification_report
```

```python
data = load_breast_cancer()
x=data.data
y=data.target

print("Classes:",data.target_names)

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3)
```

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(x_train,y_train)
pred_lr = lr.predict(x_test)
print("Logistic TestAcurracy:",accuracy_score(y_test,pred_lr))
pred_lr_train = lr.predict(x_train)
print("Logistic TrainAccuracy:",accuracy_score(y_train,pred_lr_train))
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
pred_dt = dt.predict(x_test)
print("Decision Tree TestAcurracy:",accuracy_score(y_test,pred_dt))
pred_dt_train = dt.predict(x_train)
print("Decicsion TrainAccuracy:",accuracy_score(y_train,pred_dt_train))
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
pred_knn = knn.predict(x_test)
print("KNN Tree TestAcurracy:",accuracy_score(y_test,pred_knn))
pred_knn_train = knn.predict(x_train)
print("TrainAccuracy:",accuracy_score(y_train,pred_knn_train))
```

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
pred_nb = nb.predict(x_test)
print("  GaussianN Naive Bayes Tree TestAcurracy:",accuracy_score(y_test,pred_nb))
pred_nb_train = nb.predict(x_train)
print(" GaussianN Naive Bayes TrainAccuracy:",accuracy_score(y_train,pred_nb_train))
```

```python
nn = MLPClassifier ()
nn.fit(x_train,y_train)
pred_nn = nn.predict(x_test)
print("Neural Network Tree TestAcurracy:",accuracy_score(y_test,pred_nn))
pred_nn_train = nn.predict(x_train)
print("Neural Network TrainAccuracy:",accuracy_score(y_train,pred_nn_train))
```

### Confusion Matrix

```python
cm_lr = confusion_matrix (y_test , pred_lr)
print("\nLogistic Regression Confusion Matrix :")
print(cm_lr)
TN1 , FP1 ,FN1 , TP1 = cm_lr.ravel()
type1_lr = FP1 / (FP1 + TN1)
type2_lr = FN1 / (FN1 + TP1)

print("FP1 (Type-1):",FP1)
print("FN1 (Type-2:)",FN1)
print("Type-1 Error Rate in %:",type1_lr*100)
print("Type-2 Error Rate in %:",type2_lr*100)

precision = precision_score(y_test,pred_lr)
recall = recall_score(y_test , pred_lr)
f1 = f1_score(y_test , pred_lr)
roc = roc_auc_score(y_test , pred_lr)


print("Precision:",precision)
print("Recall:",recall)
print("f1:",f1)
print("Roc:",roc)
```
